5) Utilizando la tabla "Match": 
- Obtén qué 2 equipos se han enfrentado más veces (con los ids de los equipos es suficiente)
- Obtén que 2 equipos se han enfrentado 11 veces (con los ids de los equipos es suficiente)

```python
%%sql

SELECT Team_1, Team_2
FROM Match
LIMIT 10
```

```python

%%sql

SELECT CASE WHEN Team_1 > Team_2
THEN Team_1 || "-" || Team_2 
ELSE Team_2 || "-" || Team_1
END AS Parejas_De_Equipos, COUNT(*) AS Veces_Jugadas
FROM Match
GROUP BY Parejas_De_Equipos
ORDER BY Veces_Jugadas DESC
LIMIT 10
```

10) En la tabla 'Umpire' puedes encontrar información de los árbitros, incluyendo su país de origen.
Extrae el listado de aquellos árbitros que hayan pitado algún partido en estadios de su país de origen.

```python
#Compruebo primero cómo es la tabla Umpire
%%sql
SELECT *
FROM Umpire
```

```python
#query para probar la unión de tablas. Se repiten tantos arbitros como ciudades en las que haya estado dentro de su país de origen. Pongo limit por comodidad visual, pero la consulta
#inicial ha sido completa.
%%sql
SELECT Umpire_Name as nombre_arbitro, Umpire_Country as pais_arbitro, Country_Name as nombre_pais, Venue_Name as estadio 
FROM Venue V
LEFT JOIN City C on V.City_Id=C.City_Id
LEFT JOIN Country K on C.Country_id=K.Country_Id
INNER JOIN Umpire U on K.Country_Id=U.Umpire_Country
LIMIT 50

```

```python
# Como lo que nos piden es el listado de árbitros que hayan pitado en estadios de su país de origen, agrupamos por árbitro, en concreto, por el nombre del árbitro.
%%sql
SELECT Umpire_Name as nombre_arbitro, Umpire_Country as pais_arbitro, Country_Name as nombre_pais, Venue_Name as estadio 
FROM Venue V
LEFT JOIN City C on V.City_Id=C.City_Id
LEFT JOIN Country K on C.Country_id=K.Country_Id
INNER JOIN Umpire U on K.Country_Id=U.Umpire_Country
GROUP BY nombre_arbitro
```

11) Lista el nombre de todos los estadios en los que haya jugado el equipo 'Mumbai Indians'

```python
# Compruebo primero cómo se quedaría la tabla y si los joins funcionan
%%sql
SELECT *
FROM 
  Match M
LEFT JOIN Team T on M.Team_1=T.Team_Id
LEFT JOIN Venue V on M.Venue_Id=V.Venue_Id
WHERE Team_Name== "Mumbai Indians"
LIMIT 10
```

```python
# Selecciono las columnas que me interesan 
%%sql
SELECT DISTINCT(Venue_Name)
FROM 
  Match M
LEFT JOIN Team T on M.Team_1=T.Team_Id
LEFT JOIN Venue V on M.Venue_Id=V.Venue_Id
WHERE Team_Name== "Mumbai Indians"
```

```python
# Hago el mismo join con Team_2, y mediante UNION uno las dos
%%sql
SELECT DISTINCT(Venue_Name) as nombre_estadios, Team_Name as equipo_participante
FROM 
  Match M
LEFT JOIN Team T on M.Team_1=T.Team_Id
LEFT JOIN Venue V on M.Venue_Id=V.Venue_Id
WHERE Team_Name= "Mumbai Indians"

UNION

SELECT DISTINCT(Venue_Name), Team_Name 
FROM 
  Match M
LEFT JOIN Team T on M.Team_2=T.Team_Id
LEFT JOIN Venue V on M.Venue_Id=V.Venue_Id
WHERE Team_Name= "Mumbai Indians"
```
